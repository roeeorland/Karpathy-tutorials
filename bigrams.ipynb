```python
with open ('names.txt', 'r') as fp:
    names = fp.read()
```

```python
names
```

```python
names = names.lower().splitlines()
```

```python
bigram_dict = {}
for name in names:
    name = '<>''.'
    for ch1, ch2 in zip(name, name[1:]):
        bigram = (ch1, ch2)
        bigram_dict[bigram] = bigram_dict.get(bigram, 0) + 1
    
```

```python
max(bigram_dict.values())
```

```python
bigrams_list = [(key, value) for key, value in bigram_dict.items()]
```

```python
bigrams_list.sort(key=lambda x: x[1], reverse=True)
```
